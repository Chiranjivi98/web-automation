```python
import requests
from bs4 import BeautifulSoup
```

```python
url="https://www.midsouthshooterssupply.com/dept/reloading/primers?currentpage=1"
page=requests.get(url)
print(page) # url correct if responce is 200 
```

```python
soup=BeautifulSoup(page.text,'html.parser')
data=[]
data_iterator=iter(soup.find_all('li'))
while True:
    try:
        PRICE=next(data_iterator).text
        Title=next(data_iterator).text
        Stocks=next(data_iterator).text
        Manufacturer=next(data_iterator).text
        data.append((PRICE,Title,Stocks,Manufacturer))
    except StopIteration:
        break
data.sort(key=lambda row:row[1],reverse=True)
```

```python
import texttable as tt
```

```python
table=tt.Texttable()
table.add_rows([(None,None,None,None)]+data)
table.set_cols_align(('c','c','c','c'))
table.header(('PRICE','Title','Stock status','Manufacturer'))

print(table.draw())
```
